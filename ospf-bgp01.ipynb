```python
# dev-test-host01_IP
host01_ip = "172.16.0.10"
# dev-test-host02_IP
host02_ip = "172.16.1.10"
# dev-test-host03_IP
host03_ip = "172.16.2.10"
# 
app = "http"
flow1 = HeaderConstraints(srcIps=host01_ip,
                         dstIps=host02_ip,
                         applications=[app])
flow2 = HeaderConstraints(srcIps=host01_ip,
                         dstIps=host03_ip,
                         applications=[app])
flow3 = HeaderConstraints(srcIps=host02_ip,
                         dstIps=host03_ip,
                         applications=[app])
```

```python
traceroute_ans1 = bfq.traceroute(
                                headers=flow1,
                                startLocation="dev-test-host01"
                               ).answer()
display_html(traceroute_ans1.frame())
```

```python
traceroute_ans2 = bfq.traceroute(
                                headers=flow2,
                                startLocation="dev-test-host01"
                               ).answer()
display_html(traceroute_ans2.frame())
```

```python
traceroute_ans3 = bfq.traceroute(
                                headers=flow3,
                                startLocation="dev-test-host02"
                               ).answer()
display_html(traceroute_ans3.frame())
```
